1. Merge the datasets Customers, Product Hierarchy and Transactions as Customer_Final. Ensure 
tokeep all customers who have done transactions with us and select the join type accordingly.

```python
customer_Trans=pd.merge(left=customer,right=Transaction,left_on='customer_Id',right_on='cust_id',how='inner',indicator=True)
customer_Trans
```

```python
customer_final=pd.merge(left=customer_Trans,right=products,left_on='prod_cat_code',right_on='prod_cat_code',how='inner')
customer_final
```

5.Analyze which product categories are more popular among females vs male customers

```python
#MALE
X=customer_final.loc[customer_final['Gender']=='M']
Y=X.groupby(['prod_cat'])['total_amt'].sum()
Z=Y.nlargest(1)
display("most popular category in male customer: ",Z)
#FEMALE
P=customer_final.loc[customer_final['Gender']=='F']
Q=P.groupby(['prod_cat'])['total_amt'].sum()
R=Q.nlargest(1)
display('Most popular category in Female customer:',R)
```

6. Which City code has the maximum customers and what was the percentage of customers from 
that city?

```python
count=customer_final.groupby(['city_code'])['cust_id'].count().nlargest(1)
print("Max customer",count)
percentage=(customer_final.groupby(['city_code'])['cust_id'].count())*100/(customer_final['cust_id'].count())
print("percentage customers of every city",percentage)
```

7.Which store type sells the maximum products by value and by quantity?

```python
#by value
max_store=customer_final.groupby(['Store_type'])['total_amt'].sum().nlargest(1)
print("Max products by values",max_store)
#by quantity
max_quantity=customer_final.groupby(['Store_type'])['prod_cat_code'].count().nlargest(1)
print("max products by quantity",max_quantity)
```

```python
#calculating age from date of birth
cury=pd.to_datetime("today").year
doby=pd.DatetimeIndex(customer_final['DOB']).year
customer_final['age']=cury-doby
customer_final['age']

```

```python
#calculating total amount spent on electronics and books for customers whose age is between 25 and 35
df=customer_final.loc[(customer_final['age']>=25) & (customer_final['age']<=35)]
tf=df.loc[(df['prod_cat']=='Electronics') | (df['prod_cat']=='Books')]['total_amt'].sum()
print("Total amount spent:",tf)
```

```python
#changing the formate  of transaction date
customer_final['tran_date']=pd.to_datetime(df['tran_date'],dayfirst=True).dt.strftime("%d-%m-%Y")
```

```python
# calculting total amount between 1st jan, 2014 and 1st march, 2014
df=customer_final.loc[(customer_final['age']>=25) & (customer_final['age']<=35)]
total=df.loc[(df['tran_date']>'01-01-2014') & (df['tran_date']<'01-03-2014')]['total_amt'].sum()
print("Total_amount spent:",total)
```
